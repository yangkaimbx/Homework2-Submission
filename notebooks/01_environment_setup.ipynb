## Step 5: Make Your First API Call

Let's test the connection with a quick Week 2-themed question.

```python
import time

print("=" * 60)
print("🧪 FIRST API CALL — Week 2 Orientation")
print("=" * 60)
print()

prompt = (
    "In exactly 3 bullet points, describe what you will learn in "
    "a course week covering: Transformer architecture, LLM pretraining "
    "data pipelines, tokenization, and alignment techniques."
)

start = time.time()
response = client.generate(
    prompt=prompt,
    system="You are a concise ML course assistant. Reply in 3 bullet points max.",
    max_tokens=300,
    temperature=0.7,
)
elapsed = time.time() - start

if "error" in response:
    print(f"❌ Error: {response['error']}")
    print("   Check your API key / Ollama connection")
else:
    print(f"✅ Response received in {elapsed:.1f}s")
    tracker.add_call(response)
    print(format_response(response, verbose=True))
```
