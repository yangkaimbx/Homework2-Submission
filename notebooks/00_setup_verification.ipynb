## 5: Check System Dependencies (Tesseract & ffmpeg)

````python
import shutil
import subprocess

print("=" * 60)
print("🔧 SYSTEM DEPENDENCIES CHECK")
print("=" * 60)

# Tesseract binary
tesseract_path = shutil.which('tesseract')
if tesseract_path:
    try:
        result = subprocess.run(['tesseract', '--version'], capture_output=True, text=True)
        version_line = result.stderr.split('\n')[0] if result.stderr else result.stdout.split('\n')[0]
        print(f"✅ Tesseract found: {version_line}")
    except Exception:
        print(f"✅ Tesseract found at: {tesseract_path}")
else:
    print("❌ Tesseract NOT found")
    print("   macOS:  brew install tesseract")
    print("   Ubuntu: sudo apt install tesseract-ocr")
    print("   Windows: https://github.com/tesseract-ocr/tesseract/releases")

print()

# ffmpeg (required by faster-whisper)
ffmpeg_path = shutil.which('ffmpeg')
if ffmpeg_path:
    print(f"✅ ffmpeg found at: {ffmpeg_path}")
else:
    print("❌ ffmpeg NOT found")
    print("   macOS:  brew install ffmpeg")
    print("   Ubuntu: sudo apt install ffmpeg")
    print("   Windows: https://ffmpeg.org/download.html")

print()

# poppler (required by pdf2image on macOS/Linux)
pdftoppm_path = shutil.which('pdftoppm')
if pdftoppm_path:
    print(f"✅ poppler (pdftoppm) found at: {pdftoppm_path}")
else:
    print("❌ poppler NOT found  (needed by pdf2image)")
    print("   macOS:  brew install poppler")
    print("   Ubuntu: sudo apt install poppler-utils")
````

## 6: Check Environment Variables

````python
from dotenv import load_dotenv
import os

print("=" * 60)
print("🔑 ENVIRONMENT VARIABLES CHECK")
print("=" * 60)

# Load .env from parent directory
env_path = os.path.join('..', '.env')
load_dotenv(env_path)

api_key = os.getenv('ANTHROPIC_API_KEY')
if api_key:
    print(f"✅ ANTHROPIC_API_KEY found ({len(api_key)} chars, starts with: {api_key[:7]}...)")
else:
    print("⚠️  ANTHROPIC_API_KEY not found")
    print("   OK if you're using Ollama (Path B)")
    print("   Required for Claude (Path A or C)")
    print("   → Copy .env.example to .env and add your key")

hf_token = os.getenv('HF_TOKEN') or os.getenv('HUGGINGFACE_TOKEN')
if hf_token:
    print(f"✅ HF_TOKEN found ({len(hf_token)} chars)")
else:
    print("⚠️  HF_TOKEN not found, please set it from the HuggingFace website: https://huggingface.co/settings/tokens")
    print("   Optional — needed only for gated HuggingFace models")
    print("   Public models (bert-base-uncased, gpt2, t5-small) work without it")
````

## 7: Check Ollama (Path B/C)

````python
import requests

print("=" * 60)
print("🦙 OLLAMA CHECK")
print("=" * 60)

try:
    resp = requests.get('http://localhost:11434/api/tags', timeout=5)
    if resp.status_code == 200:
        models = resp.json().get('models', [])
        if models:
            print("✅ Ollama is running")
            print(f"   Available models ({len(models)}):")
            for m in models:
                flag = " ← Week 2 default" if m['name'] == 'qwen3.5:27b' else ""
                print(f"     - {m['name']}{flag}")
            if not any(m['name'] == 'qwen3.5:27b' for m in models):
                print()
                print("   💡 Recommended: ollama pull qwen3.5:27b")
        else:
            print("⚠️  Ollama running but no models found")
            print("   Run: ollama pull qwen3.5:27b")
except Exception:
    print("⚠️  Ollama not running")
    print("   OK if you're using Claude API (Path A)")
    print("   To start: ollama serve")
````
